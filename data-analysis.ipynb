```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
full_dataset = pd.read_csv('wine-clustering.csv')
print(full_dataset.columns)
full_dataset.describe()
```

```python
fig, axes = plt.subplots(1, len(full_dataset.columns), figsize=(30,10))

for i in np.arange(len(full_dataset.columns)):

    title = full_dataset.columns[i]
    axes[i].set_title(title)
    sns.histplot(full_dataset, x=title, ax=axes[i])

plt.show()
```

```python
corr_var = full_dataset.corr(method ='pearson')
sns.heatmap(corr_var)
plt.show()
corr_var
```

The highest correlation is between Total_Phenols and Flavanoids (0.864564 out of 1)
